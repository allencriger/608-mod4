```python
file = open('text_about_python.txt')

read_data = file.read()

unique_word = read_data.split()

print('Total Words in "text_about_python" document is:', len(unique_word))
print(" ")

from collections import Counter

counter = Counter(unique_word)

for word, count in sorted(counter.items()):
    print(f'{word:<20}{count}')

```

```python
filename = 'text_about_python.txt'

file = open(filename, 'rt')

text = file.read()

file.close()

words = text.split()

words = [word.lower() for word in words]

import string

table = str.maketrans('','', string.punctuation)

stripped = [w.translate(table) for w in words]

from collections import Counter

counter = Counter(stripped)

for clean_list, count in sorted(counter.items()):
    print((f'{clean_list:<20}{count}')) 
```
